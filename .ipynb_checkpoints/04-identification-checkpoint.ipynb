<h2 style='font-size:30px'> Estimand & Conditional Ignorability
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Em Inferência Causal, existem os conceitos de <i> estimand</i>, <i> estimator</i> e <i> estimate</i> que podem ser confundidos uns pelos outros.
            <ul> 
                <li>
                    <i> Estimand:</i> Uma determinada métrica que desejamos estimar (ATE, ATC...)
                </li>
                <li>
                    <i> Estimator:</i> Método/Técnica usada na computação do estimand.
                </li>
                <li>
                    <i> Estimate:</i> Seria o valor do estimand obtido por meio do estimator.
                </li>
            </ul>
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Estimands</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A Inferência Causal lida com dois tipos de estimands:
            <ul> 
                <li>
                    <i> Causal Estimand:</i> Qualquer métrica que envolva potential outcomes e o do-operator.
                </li>
                <li>
                    <i> Statistical Estimand:</i> Qualquer métrica baseda em distribuições de variáveis observadas, que não contenha potential outcomes ou do-operator.
                </li>
            </ul>
        </li>
        <li>
            Podemos considerar um causal estimand como <i>identificável</i>, caso consiga ser expresso como um statistical estimand. Isso significa que poderíamos usar técnicas de primeiro degrau em dados observacionais para mensurar métricas causais!
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Ignorability</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Como o curso já havia pontuado, o estado da arte de um estudo observacional causal ocorre quando as amostras de tratamento e controle têm distribuições de covariáveis equivalentes.
        </li>
        <li>
            Nessa situação, espera-se que, se a intervenção tivesse sido aplicada sobre as instâncias de controle, e não tratamento, o efeito observado teria sido o mesmo. Chamamos esse fenômeno de "Ignorabilidade". 
        </li>
    </ul>
</div>

 <center style='margin-top:20px;font-size:20px'> 
        <figure>
            $Y(0),Y(1){\perp\!\!\!\perp}T$
            <figcaption style='font-size:15px'> 
                Formalização do conceito de Ignorabilidade.
            </figcaption>
        </figure>
    </center>

 <center style='margin-top:20px;font-size:20px'> 
        <figure>
            $$
            \begin{align*}
            E[Y(1)|T=1]&=E[Y(1)|T=0]=E[Y(1)]\\
            E[Y(0)|T=0]&=E[Y(0)|T=1]=E[Y(0)]
            \end{align*}
            $$
            <figcaption style='font-size:15px'> 
            </figcaption>
        </figure>
    </center>

<h3 style='font-size:30px;font-style:italic'> Conditional Ignorability</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Como é de se supor, é praticamente impossível se alcançar a Ignorabilidade plena num estudo. Porém, pode ser que as covariáveis de tratamento e controle assumam distribuições parecidas, quando essas são condicionadas por alguma covariável $Z$.
        </li>
    </ul>
</div>

 <center style='margin-top:20px;font-size:20px'> 
        <figure>
            $$
            \begin{align*}
            E[Y(1)|T=1,Z=z]&=E[Y(1)|T=0,Z=z]=E[Y(1)|Z=z]\\
            E[Y(0)|T=0,Z=z]&=E[Y(0)|T=1,Z=z]=E[Y(0)|Z=z]
            \end{align*}
            $$
            <figcaption style='font-size:15px'> 
            </figcaption>
        </figure>
    </center>

<h3 style='font-size:30px;font-style:italic'> Adjustment Formula</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Assumindo que nossos dados possuem Ignorabilidade Condicional, poderemos considerar o $\operatorname{ATE}$ como a média das diferenças entre efeitos de tratatamento e controle para cada valor de $Z$. 
        </li>
    </ul>
</div>

 <center style='margin-top:20px;font-size:20px'> 
        <figure>
            $$
            \begin{align*}
            \operatorname{ATE}&=\mathbb{E}{[Y|\operatorname{do}(T=1)]}-\mathbb{E}{[Y|\operatorname{do}(T=0)]}\\
            &=\mathbb{E}[Y(1)]-\mathbb{E}[Y(0)]\\
            &=\mathbb{E}_{Z}[\mathbb{E}[Y(1)|Z]-\mathbb{E}[Y(0)|Z]]\\
            &=\mathbb{E}_{Z}[\mathbb{E}[Y(1)|T=1,Z]-\mathbb{E}[Y(0)|T=0,Z]]\\
            &=\mathbb{E}_{Z}[\mathbb{E}[Y|T=1,Z]-\mathbb{E}[Y|T=0,Z]]\text{ (Consistency Assumption)}
            \end{align*}
            $$
            <figcaption style='font-size:15px'> 
            </figcaption>
        </figure>
    </center>

<div> 
    <ul style='font-size:20px'> 
        <li>
            Repare que as transformações feitas acima tornam a mensuração do $\operatorname{ATE}$ identificável!
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    Adjustment Formula Premises
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Repare que a fórmula do ajuste pressupõe as condições abaixo:
            <ul> 
                <li>
                    <i>No Interference:</i> SUTVA (o efeito de uma instância não afeta os das outras).
                </li>
                <li>
                    <i> Positivity:</i> $P(T=t|Z=z)>0,\forall{z\in{Z}}$. Caso o contrário, $E[Y|T=t,Z=z]$ será indefinido.
                </li>
            </ul>
        </li>
    </ul>
</div>

        <li>

        </li>
        <li>
            A partir de agora, vamos explorar outras métricas causais baseadas na probabilidade de distribuições, ampliando o nosso repertório.

<h2 style='font-size:30px'> Backdoor & Frontdoor Adjustment
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Para nos auxiliar na computação de métricas causais, vamos introduzir o conceito de Backdoor e Frontdoor Adjustment.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Backdoor Adjustment</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Entendemos como back door paths todos os caminhos em formato de fork que geram confundimento na mensuração do efeito do tratamento na variável.
        </li>
        <li>
            A fim de isolar o confundimento, nós estratificamos o dataset com base nos valores das variáveis confundidoras, e mensuramos o $\operatorname{ATE}$ como a média dos sub-$\operatorname{ATE}$'s de cada agrupamento (como foi visto na aula sobre Ignorability).
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> 
    <a href='https://bayesianquest.com/2023/10/01/causal-analysis-foundation-series-back-door-front-door-and-instrumentals-in-causal-analysis/'> Frontdoor Adjustment</a>
</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            O Frontdoor Adjustment lida com situações em que certos confundidores não conseguem ser mensurados ou inclusos no nosso dataset.
        </li>
        <li>
            Para que consigamos estimar o efeito, mesmo sem esssas variáveis em mãos, tentamos localizar um mediador $Z$, que intermedeia todos os caminhos entre $X$ e $Y$.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> Do-Calculus
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Do-Calculus consiste numa coleção de propriedades em Grafos Causais que podem ser úteis em tornar métricas causais identificáveis.
        </li>
        <li>
            Na elaboração dos teoremas, usamos uma notação especial para grafos que será especificada abaixo.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    Notações de $G$
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            $G_{\overline{X}}$: Representação de $G$ em que todos os edges que entram em $X$ são removidos.
        </li>
        <li>
            $G_{\underline{X}}$: Representação de $G$ em que todos os edges saindo de $X$ são removidos.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Rule 1</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A primeira regra do Do-Calculus afirma que:
        </li>
    </ul>
</div>

 <center style='margin-top:20px;font-size:20px'> 
        <figure>
            $\displaystyle P(y|z,\operatorname{do}(x),w)=P(y|\operatorname{do}(x),w)\iff{(Y{\perp\!\!\!\perp}Z|W,X)}\text{ in }G_{\overline{X}}$
            <figcaption style='font-size:15px'> 
            </figcaption>
        </figure>
    </center>

<div> 
    <ul style='font-size:20px'> 
        <li>
            Observe que isso faz total sentido! Sendo $Y{\perp\!\!\!\perp}Z$, podemos desconsiderar $Z$ na computação de probabilidades de $Y$.
        </li>
        <li>
            Essa regra somente é válida em $G_{\overline{X}}$ pois estamos intervindo em $X$ ($\operatorname{do}(x)$). 
        </li>
    </ul>
</div>
