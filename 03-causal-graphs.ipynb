<h2 style='font-size:30px'> What are Causal DAGs?
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Causal Directed Acyclic Graphs são graos usados para esquematizar a relação lógica entre variáveis de nossos estudos.
        </li>
        <li>
            A Inferência Causal presume algumas condições que devem ser satisfeitas por todo DAG:
            <ul> 
                <li>
                    <i>Directed:</i> As ligações devem ser unidirecionais, significando que variáveis não podem se causar mutualmente.
                </li>
                <li>
                    <i>Acyclic:</i> Um caminho de edges não pode retornar a um nó já percorrido. 
                </li>
            </ul>
        </li>
    </ul>
</div>
