<h2 style='font-size:30px'> What are Causal DAGs?
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Causal Directed Acyclic Graphs são graos usados para esquematizar a relação lógica entre variáveis de nossos estudos.
        </li>
        <li>
            A Inferência Causal presume algumas condições que devem ser satisfeitas por todo DAG:
            <ul> 
                <li>
                    <i>Directed:</i> As ligações devem ser unidirecionais, significando que variáveis não podem se causar mutualmente.
                </li>
                <li>
                    <i>Acyclic:</i> Um caminho de edges não pode retornar a um nó já percorrido, criando asim um ciclo no grafo. 
                </li>
            </ul>
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Fork</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Padrão em que uma variável $Z$ é a causa mútua de outras duas ($X$ e $Y$), sem que estas tenham relação entre si.
        </li>
        <li>
            Nesse tipo de geração de dados, pode ocorrer uma relação espúria entre $X$ e $Y$ ocasionada pelos valores de $Z$, sendo este considerado um <i>Confounder</i>.
        </li>
        <li>
            Essa associação espúria é melhor denunciada ao analisarmos $X$ e $Y$ tendo $Z$ um valor constante.
        </li>
    </ul>
</div>
