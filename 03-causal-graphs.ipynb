<h2 style='font-size:30px'> What are Causal DAGs?
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Causal Directed Acyclic Graphs são grafos usados para esquematizar a relação lógica entre variáveis de nossos estudos.
        </li>
        <li>
            A Inferência Causal presume algumas condições que devem ser satisfeitas por todo DAG:
            <ul> 
                <li>
                    <i>Directed:</i> As ligações devem ser unidirecionais, significando que variáveis não podem se causar mutualmente.
                </li>
                <li>
                    <i>Acyclic:</i> Um caminho de edges não pode retornar a um nó já percorrido, criando asim um ciclo no grafo. 
                </li>
            </ul>
        </li>
        <li>
            Num grafo causal, consideramos haver um fluxo de informação causal em caminhos diretos. Todos os outros tipos são reconhecidos como espúrios.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> Graph Patterns
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Nessa aula, estudamos os principais padrões em formatos de grafos em estudos causais. 
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Fork</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Padrão em que uma variável $Z$ é a causa mútua de outras duas ($X$ e $Y$), sem que estas tenham relação entre si.
        </li>
        <li>
            Nesse tipo de geração de dados, pode ocorrer uma relação espúria entre $X$ e $Y$ ocasionada pelos valores de $Z$, sendo este considerado um <i>Confounder</i>.
        </li>
        <li>
            Essa associação espúria é melhor denunciada ao analisarmos $X$ e $Y$ tendo $Z$ um valor constante.
        </li>
    </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'>
        <img src='img/03-fork.png'>
        $$
        \begin{align*}
        X{\perp\!\!\!\perp}_{G} Y\\
        X\not\perp\!\!\!\perp_{G} Y | Z
        \end{align*}
        $$
        <figcaption style='font-size:15px'>
            Exemplificação de um processo de geração de dados em fork. Como exemplo, poderíamos imaginar uma análise de número de afogamentos e vendas de sorvete em dias de diferentes temperaturas.
        </figcaption>
    </center>
</figure>

<h3 style='font-size:30px;font-style:italic'> Chain</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Processo em que variáveis se causam de maneira sequencial.
        </li>
        <Li>
            Segundo o curso, dado um fluxo $X\to Z\to Y$, ao condicionarmos $Z$, tornamos $X$ independente de $Y$.
        </li>
    </ul>
</div>

        X\not\perp\!\!\!\perp_{G} Y | Z

        $$

<h3 style='font-size:30px;font-style:italic'> Collider</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Nessa estrutura, temos duas variáveis $X$ e $$ que causam simultaneamente uma terceira $Z$.
        </li>
        <li>
            Por mais que $X$ e $Y$ sejam independentes no esquema, elas podem adquirir associação, ao as analisarmos condicionando $Z$.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> Blocking Paths & D-separation
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Em Inferência Causal, a <i>d-separation</i> ocorre quando duas variáveis $X$ e $Y$ se tornam independentes pelo condicionamento de uma terceira variável $Z$ no grafo.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Tipos de Bloqueio</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Em fluxos chain $X\rightarrow Z \rightarrow Y$, o condicionamento de $Z$ interrompe o processo natural de geração de $Y$ (indiretamente dependente de $X$). Com isso, $Y$ passa a se tornar independente de $X$. 
        </li>
        <li>
            Em grafos fork $X\leftarrow Z \rightarrow{Y}$, é possível haver uma relação espúria entre $X$ e $Y$. No entanto, ela pode ser eliminada ao condicionarmos $Z$.
        </li>
        <li>
            Em grafos collider, $X\rightarrow Z \leftarrow Y$, $X$ e $Y$ estão d-separadas naturalmente. Porém, elas poderão adquirir uma associação espúria, caso condicionemos $Z$.
        </li>
    </ul>
</div>
