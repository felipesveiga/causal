<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Structural Causal Model (SCM)</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Um Structural Causal Model é o principal modelo usado em cases de terceiro degrau de causalidade. 
        </li>
        <li>
            Esse consiste num grafo que mapeia o processo de geração de dados do nosso problema, o que seria a relação lógica entre as variáveis, assim como o impacto quantitativo de uma em outra. 
        </li>
    </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'> 
            <img src='img/02-scm.png'>
        <figcaption style='font-size:15px'>
            Representação de um SCM.
        </figcaption>
    </center>
</figure>

<div> 
    <ul style='font-size:20px'> 
        <li>
            A relação quantitativa das variáveis é expressa num sistema de equações lineares. Seus componentes podem ser tanto variáveis mapeadas pelo SCM, quanto os chamados fatores idiossincráticos (variáveis não capturadas pelo modelo, ou noise), representados como $U$.
        </li>
        <li>
            Note que cada fator $U$ terá um valor diferente por indivíduo, não sendo uma constante universal para o nosso problema.
        </li>
    </ul>
</div>

<h5 style='font-size:25px;font-style:italic;text-decoration:underline'> Estimating Counterfactuals</h5>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Uma vez tendo o nosso sistema de equações montado, podemos usá-lo para simular cenários contrafactuais para nossos problemas.
        </li>
        <li>
            A estimação desse cenário pode ser dividida em 3 etapas:
            <ul> 
                <li>
                    <i> 1. Abduction</i>: Estimação de fatores idiossincráticos.
                </li>
                <li>
                    <i> 2. Action</i>: Simulação do cenário contrafactual de nosso interesse.
                </li>
                <li>
                    <i> 3. Prediction</i>: Cálculo do efeito no cenário contrafactual.
                </li>
            </ul>
        </li>
    </ul>
</div>

<h5 style='font-size:25px;font-style:italic;text-decoration:underline'> Exemplo</h5>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Supomos que estamos estudando o impacto da experiência de trabalho e posse de um diploma universitário no salário de uma pessoa, havendo montado o seguinte sistema de equações:
        </li>
    </ul>
</div>

 <center style='margin-top:20px;font-size:20px'> 
        <figure>
            $$
            \displaystyle
            \begin{equation}
                \begin{cases}
                    S:=65000+2500\operatorname{EX}+5000\operatorname{CD}+U_{S}\\
                    \operatorname{EX}:=10-4\operatorname{CD}+U_{\operatorname{EX}}\\
                    \operatorname{CD}:=U_{\operatorname{CD}}
                \end{cases}
            \end{equation}
            $$
            <figcaption style='font-size:15px'> 
                O walrus-operator ($:=$) especifica a dependência causal da variável à esquerda das da direita. Isso significa que não podemos reorganizar as equações, pois estaríamos declarando um processo de geração de dados não existente.
            </figcaption>
        </figure>
    </center>

<div> 
    <ul style='font-size:20px'> 
        <li>
            Dada uma pessoa com 6 anos de experiência sem diploma que recebe 81000, quanto ela receberia caso tivesse uma formação?
        </li>
    </ul>
</div>

```python
# Definição dos valores das variáveis.
S = 81000
CD = 0
EX = 6
```

```python
# Estimação dos fatores idiossincráticos.

# Note que não computamos U_CD, porque ele não será usado em nossa análise.
```

```python
U_EX = EX - (10 - 4*CD)
U_EX
```

```python
U_S = S - (65000 + 2500*EX + 5000*CD)
U_S
```

```python
# Similação do cenário contrafactual
CD_ = 1
```

```python
EX_ = 10 - 4*CD_+U_EX
EX_
```

```python
# Previsão
```

```python
S_ = 65000 + 2500 * EX_ + 5000*CD_ + U_S
S_
```

```python
# Ou seja, o invíduo estaria recebendo 5000 a menos, caso tivesse ido à faculdade, segundo nossa similação.
S_ - S
```
